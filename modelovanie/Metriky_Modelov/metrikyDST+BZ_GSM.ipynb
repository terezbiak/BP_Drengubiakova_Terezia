```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'precision_0': [0.98, 0.94, 0.92, 0.91, 0.9, 0.9],
    'precision_1': [0.92, 0.87, 0.86, 0.84, 0.82, 0.77],
    'recall_0': [0.98, 0.97, 0.97, 0.97, 0.96, 0.95],
    'recall_1': [0.91, 0.83, 0.74, 0.70, 0.66, 0.69],
    'F1_0': [0.98, 0.96, 0.95, 0.95, 0.94, 0.94],
    'F1_1': [0.91, 0.85, 0.79, 0.76, 0.73, 0.73]
}
index = [1, 2, 3, 4, 5, 6]
metrics_df = pd.DataFrame(data, index=index)

plt.figure(figsize=(7, 5))

plt.plot(metrics_df.index, metrics_df['precision_0'], marker='o', label='Presnos 0')
plt.plot(metrics_df.index, metrics_df['precision_1'], marker='o', label='Presnos 1')
plt.plot(metrics_df.index, metrics_df['recall_0'], marker='o', label='Nvratnos 0')
plt.plot(metrics_df.index, metrics_df['recall_1'], marker='o', label='Nvratnos 1')
plt.plot(metrics_df.index, metrics_df['F1_0'], marker='o', label='F1 0')
plt.plot(metrics_df.index, metrics_df['F1_1'], marker='o', label='F1 1')

for i in metrics_df.index:
    plt.text(i, metrics_df['precision_0'][i], f'{metrics_df["precision_0"][i]:.4f}', ha='center', va='bottom', fontsize=9)
    plt.text(i, metrics_df['precision_1'][i], f'{metrics_df["precision_1"][i]:.4f}', ha='center', va='bottom', fontsize=9)
    plt.text(i, metrics_df['recall_0'][i], f'{metrics_df["recall_0"][i]:.4f}', ha='center', va='bottom', fontsize=9)
    plt.text(i, metrics_df['recall_1'][i], f'{metrics_df["recall_1"][i]:.4f}', ha='center', va='bottom', fontsize=9)
    plt.text(i, metrics_df['F1_0'][i], f'{metrics_df["F1_0"][i]:.4f}', ha='center', va='bottom', fontsize=9)
    plt.text(i, metrics_df['F1_1'][i], f'{metrics_df["F1_1"][i]:.4f}', ha='center', va='bottom', fontsize=9)

plt.title('Metriky pre jednotliv hodiny (DST, BZ_GSM)')
plt.xlabel('as predpovede dopredu')
plt.ylabel('Hodnoty Metrk')
plt.xticks(index)
plt.ylim(0.6, 1.0)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig("metrikyDST+BZ1")

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'ACC': [0.9663, 0.9418, 0.9244, 0.9134, 0.9060, 0.9004],
    'MCC': [0.8938, 0.8133, 0.7516, 0.7138, 0.6909, 0.6677]
}
index = [1, 2, 3, 4, 5, 6]
metrics_df = pd.DataFrame(data, index=index)

plt.figure(figsize=(5, 5))

plt.plot(metrics_df.index, metrics_df['ACC'], marker='o', label='ACC')
plt.plot(metrics_df.index, metrics_df['MCC'], marker='o', label='MCC')


for i in metrics_df.index:
    plt.text(i, metrics_df['ACC'][i], f'{metrics_df["ACC"][i]:.4f}', ha='center', va='bottom', fontsize=9)
    plt.text(i, metrics_df['MCC'][i], f'{metrics_df["MCC"][i]:.4f}', ha='center', va='bottom', fontsize=9)


plt.title('Metriky pre jednotliv hodiny (DST, BZ_GSM)')
plt.xlabel('as predpovede dopredu')
plt.ylabel('Hodnoty Metrk')
plt.xticks(index)
plt.ylim(0.6, 1.0)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig("metrikyDST+BZ2")

plt.show()
```
