```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'precision_0': [0.98, 0.96, 0.95, 0.94, 0.92, 0.92],
    'precision_1': [0.92, 0.86, 0.83, 0.81, 0.81, 0.8],
    'recall_0': [0.98, 0.97, 0.96, 0.96, 0.96, 0.96],
    'recall_1': [0.9, 0.83, 0.78, 0.73, 0.68, 0.67],
    'F1_0': [0.98, 0.96, 0.95, 0.95, 0.94, 0.94],
    'F1_1': [0.91, 0.85, 0.8, 0.77, 0.74, 0.73]
}
index = [1, 2, 3, 4, 5, 6]
metrics_df = pd.DataFrame(data, index=index)

plt.figure(figsize=(7, 5))

plt.plot(metrics_df.index, metrics_df['precision_0'], marker='o', label='Presnosť 0')
plt.plot(metrics_df.index, metrics_df['precision_1'], marker='o', label='Presnosť 1')
plt.plot(metrics_df.index, metrics_df['recall_0'], marker='o', label='Návratnosť 0')
plt.plot(metrics_df.index, metrics_df['recall_1'], marker='o', label='Návratnosť 1')
plt.plot(metrics_df.index, metrics_df['F1_0'], marker='o', label='F1 0')
plt.plot(metrics_df.index, metrics_df['F1_1'], marker='o', label='F1 1')

for i in metrics_df.index:
    plt.text(i, metrics_df['precision_0'][i], f'{metrics_df["precision_0"][i]:.4f}', ha='center', va='bottom', fontsize=9)
    plt.text(i, metrics_df['precision_1'][i], f'{metrics_df["precision_1"][i]:.4f}', ha='center', va='bottom', fontsize=9)
    plt.text(i, metrics_df['recall_0'][i], f'{metrics_df["recall_0"][i]:.4f}', ha='center', va='bottom', fontsize=9)
    plt.text(i, metrics_df['recall_1'][i], f'{metrics_df["recall_1"][i]:.4f}', ha='center', va='bottom', fontsize=9)
    plt.text(i, metrics_df['F1_0'][i], f'{metrics_df["F1_0"][i]:.4f}', ha='center', va='bottom', fontsize=9)
    plt.text(i, metrics_df['F1_1'][i], f'{metrics_df["F1_1"][i]:.4f}', ha='center', va='bottom', fontsize=9)

plt.title('Metriky pre jednotlivé hodiny (DST, BZ_GSM, KP)')
plt.xlabel('Čas predpovede dopredu')
plt.ylabel('Hodnoty Metrík')
plt.xticks(index)
plt.ylim(0.6, 1.0)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig("metrikyDST+BZ+KP1_Transform")

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'ACC': [0.9643, 0.9407, 0.9238, 0.9126, 0.9058, 0.9007],
    'MCC': [0.8869, 0.8105, 0.7550, 0.7159, 0.6881, 0.6710]
}
index = [1, 2, 3, 4, 5, 6]
metrics_df = pd.DataFrame(data, index=index)

plt.figure(figsize=(5, 5))

plt.plot(metrics_df.index, metrics_df['ACC'], marker='o', label='ACC')
plt.plot(metrics_df.index, metrics_df['MCC'], marker='o', label='MCC')


for i in metrics_df.index:
    plt.text(i, metrics_df['ACC'][i], f'{metrics_df["ACC"][i]:.4f}', ha='center', va='bottom', fontsize=9)
    plt.text(i, metrics_df['MCC'][i], f'{metrics_df["MCC"][i]:.4f}', ha='center', va='bottom', fontsize=9)


plt.title('Metriky pre jednotlivé hodiny (DST, BZ_GSM, KP)')
plt.xlabel('Čas predpovede dopredu')
plt.ylabel('Hodnoty Metrík')
plt.xticks(index)
plt.ylim(0.6, 1.0)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig("metrikyDST+BZ+KP2_Transform")

plt.show()
```
