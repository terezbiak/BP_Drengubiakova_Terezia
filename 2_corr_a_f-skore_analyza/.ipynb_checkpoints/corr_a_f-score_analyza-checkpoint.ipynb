### Import knižníc

```python
import numpy as np
import pandas as pd
import datetime
import math
from scipy import stats
import seaborn as sns
import matplotlib.dates as md
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
from matplotlib.ticker import MaxNLocator
from sklearn.feature_selection import SelectKBest, f_regression
```

### Korelačná analýza

```python
# Načítanie dát
df = pd.read_csv('omni_full.csv')
df = df.drop(columns=['Unnamed: 0', "time1", 'BZ_GSM-1','BZ_GSM-2','BZ_GSM-3','BZ_GSM-4','BZ_GSM-5','BZ_GSM-6',
                     'BZ_GSM-7','BZ_GSM-8','BZ_GSM-9','BZ_GSM-10','BZ_GSM-11','BZ_GSM-12','BZ_GSM-13','BZ_GSM-14',
                     'BZ_GSM-15','BZ_GSM-16','BZ_GSM-17','BZ_GSM-18','BZ_GSM-19','BZ_GSM-20','BZ_GSM-21','BZ_GSM-22','BZ_GSM-23','BZ_GSM-24'])
# Výpočet korelačnej matice
korelacna_matica = df.corr()

total_attributes = len(korelacna_matica.columns)
mid_index = total_attributes // 2  # Nájdenie stredného indexu na rozdelenie

# Prvá polovica atribútov
first_half = korelacna_matica.columns[:mid_index]

# Druhá polovica atribútov
second_half = korelacna_matica.columns[mid_index:]

# Vytvorenie heatmapy pre prvú polovicu atribútov v korelácii s DST
plt.figure(figsize=(15, 5))
sns.heatmap(korelacna_matica.loc[['DST'], first_half], annot=True, fmt=".2f", cmap="coolwarm")
plt.title('Korelačná heatmapa pre prvú polovicu atribútov s DST')
plt.savefig("corr1")
plt.show()

# Vytvorenie heatmapy pre druhú polovicu atribútov v korelácii s DST
plt.figure(figsize=(15, 5))
sns.heatmap(korelacna_matica.loc[['DST'], second_half], annot=True, fmt=".2f", cmap="coolwarm")
plt.title('Korelačná heatmapa pre druhú polovicu atribútov s DST')
plt.savefig("corr2")
plt.show()
```

### F-skóre analýza

```python
# Načítanie a príprava dát
df = pd.read_csv('omni_full.csv')

#odstránenie atribútu time1
df = df.drop(columns=['time1'])

#nahradenie chýbajúcich hodnôt mediánom
df.fillna(df.median(), inplace=True)

#Cieľová premenná DST sa uloží do premennej y
y = df['DST'].values

#Vstupné atribúty sa uložia do premennej X (všetky stĺpce okrem DST)
X = df.drop(columns=['DST']).values

#Názvy atribútov (okrem DST) sa uložia do premennej feature_labels
feature_labels = df.drop(columns=['DST']).columns.values

#Použije sa model SelectKBest s funkciou skóre f_regression na výber 10 najlepších atribútov (numFeatures).
numFeatures = 10
fSelect_model = SelectKBest(score_func=f_regression, k=numFeatures)
X_fSelect = fSelect_model.fit_transform(X, y)
scores = fSelect_model.scores_
scores[np.isnan(scores)] = 0  
order = np.argsort(scores)
ordered_feature_labels = feature_labels[order]
y_pos = np.arange(len(feature_labels))

#vykreslenie grafu
plt.figure(figsize=(10, 8))
plt.barh(y_pos[-numFeatures:], scores[order][-numFeatures:], align='center', color="#EFBF7B")
plt.yticks(y_pos[-numFeatures:], ordered_feature_labels[-numFeatures:], fontsize=12)
plt.xlabel('F-skóre', fontsize=15)
plt.title('Top 10 atribútov na základe F-skóre', fontsize=15)
plt.grid(axis='x', linestyle='--', alpha=0.6) 
plt.savefig("KBest")
plt.show()


selected_feature_labels = ordered_feature_labels[-numFeatures:]
selected_scores = scores[order][-numFeatures:]

#výpis hodnôt
print("Selected features and their F-scores:")
for label, score in zip(selected_feature_labels, selected_scores):
    print(f"{label}: {score}")
```
